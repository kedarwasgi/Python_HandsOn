### 3. Write a fun which will take input as a dict and output as a list of values.

```python
def fun_DictToList(d):
    if type(d)==dict:
        l=[]
        for i in d.values():
            if type(i)==dict:
                for j in i.values():
                    l.append(j)
            else:
                l.append(i)
    return l


fun_DictToList({"abc":123,"def":456, "ghi":{"a":789}})
```

### 4. Write a fun which will take another function as an input and return an output.

```python
def fun_circumf(diameter):
    circumf=3.14*diameter
    return round(circumf,2)
    
def fun_area(fun):
    circumf=fun
    area= circumf**2/(4*3.14)
    return round(area,2)
    
fun_area(fun_circumf(5))
```

### 5. Write a fun to take multiple lists as an input and output as concatenated list.

```python
def fun_list(*lists):
    
    l = []
    
    for i in lists:
        l = l + i
           
    return l

fun_list([12,5,3,4,7],[2,8],[5,15,20])
```
